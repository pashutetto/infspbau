Рассмотрим плотность распределения экспериментальных данных, полученных при измерении сопротивления резисторов

```python
%matplotlib inline

import pandas
import random

import matplotlib
import numpy
import matplotlib.pyplot as plt

df = pandas.read_csv("resistors.csv")

df['resistance'].plot(kind='bar')

print(df.resistance.describe())

df.plot.kde()
```

Проверим данные с помощью критерия Колмогорова-Смирнова


```python
from scipy import stats

d = df['resistance']

print(stats.kstest(d , 'norm', (df.mean(), df.std())))
```

Заметим, что pvalue достаточно большое, поэтому можно утверждать, что распределение может быть нормальным
