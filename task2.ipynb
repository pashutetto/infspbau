```python
import math

ITERATIONS = 20

def my_ln(x):
    """
    Вычисление натурального логарифма 1+x при помощи частичного суммирования
    ряда Тейлора для окрестности 0
    """
    x_pow = x
    multiplier = 1
    partial_sum = x
    for n in range(1, ITERATIONS):
        x_pow *= x  # В цикле постепенно считаем степень
        multiplier *= -1  # (-1)^n 
        partial_sum += x_pow * multiplier/(n+1)
    
    return partial_sum

help(math.log1p)
help(my_ln)

print(math.log1p(0.4))
print(my_ln(0.4))
```

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ['svg',  'pdf']
%matplotlib inline


import matplotlib.pyplot as plt
import numpy as np

vs = np.vectorize(my_ln)
print(my_ln, vs)

angles = np.r_[-1.00:1.00:0.01]
plt.plot(angles, np.log1p(angles), linewidth=3.0, color='cyan')
plt.plot(angles, vs(angles), linewidth=1.0, color='black')
plt.show()
```
